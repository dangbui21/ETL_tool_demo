```python
from bs4 import BeautifulSoup
```

```python
html_input = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Example HTML Input</title>
</head>
<body>
    <div class="content">
        <p>This is a paragraph.</p>
        <p class="highlight">This is another paragraph with a class.</p>
        <div>
            <a href="https://www.example.com">Link 1</a>
            <a href="https://www.example.com" class="link">Link 2</a>
        </div>
        <p>Yet another paragraph.</p>
    </div>
</body>
</html>
"""
```

```python
# Tạo đối tượng BeautifulSoup từ chuỗi HTML
soup = BeautifulSoup(html_input, 'html.parser')
```

```python
# Ví dụ về thuộc tính 'text'
print("Ví dụ về thuộc tính 'text':")
paragraph = soup.find('p')
print(paragraph.text)  # In ra nội dung của thẻ <p>
```

```python
# Ví dụ về thuộc tính 'parent'
print("\nVí dụ về thuộc tính 'parent':")
paragraph = soup.find('p')
parent_tag = paragraph.parent  # Trả về thẻ cha của thẻ <p>
print(parent_tag.name)
```

```python
# Ví dụ về thuộc tính 'parents'
print("\nVí dụ về thuộc tính 'parents':")
paragraph = soup.find('p')
for parent_tag in paragraph.parents:  # Duyệt qua tất cả các thẻ cha của thẻ <p>
    print(parent_tag.name)
```

```python
# Ví dụ về thuộc tính 'string'
print("\nVí dụ về thuộc tính 'string':")
paragraph = soup.find('p')
paragraph_text = paragraph.string  # Trả về nội dung văn bản của thẻ <p>
print(paragraph_text)
```

```python
# Ví dụ về phương thức 'find(tag)'
print("Ví dụ về phương thức 'find(tag)':")
paragraph = soup.find('p')  # Tìm thẻ đầu tiên <p> trong tài liệu
print(paragraph)
```

```python
# Ví dụ về phương thức 'find_all(tag)'
print("\nVí dụ về phương thức 'find_all(tag)':")
paragraphs = soup.find_all('p')  # Tìm tất cả các thẻ <p> trong tài liệu
for p in paragraphs:
    print(p)
```

```python
# Ví dụ về phương thức 'select(selector)'
print("\nVí dụ về phương thức 'select(selector)':")
links = soup.select('a.link') # Chọn tất cả các thẻ <a> có class="link"
for link in links:
    print(link)
```

```python
# Ví dụ về phương thức 'find_parent(name, attrs)'
print("\nVí dụ về phương thức 'find_parent(name, attrs)':")
paragraph = soup.find('p')
parent_div = paragraph.find_parent('div', class_='content')  # Tìm thẻ cha đầu tiên <div class="content">
print(parent_div)
```

```python
# Ví dụ về phương thức 'find_next_sibling(name, attrs)'
print("\nVí dụ về phương thức 'find_next_sibling(name, attrs)':")
paragraph = soup.find('p')
next_paragraph = paragraph.find_next_sibling('p')  # Tìm thẻ p kế tiếp
print(next_paragraph)
```

```python
# Ví dụ về phương thức 'find_previous_sibling(name, attrs)'
print("\nVí dụ về phương thức 'find_previous_sibling(name, attrs)':")
next_paragraph = paragraph.find_previous_sibling('p')  # Tìm thẻ p trước đó
print(next_paragraph)
```

```python
# Ví dụ về phương thức 'prettify()'
print("\nVí dụ về phương thức 'prettify()':")
print(soup.prettify())  # In ra tài liệu được căn chỉnh đẹp hơn
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
df = pd.read_csv("data_gop2.csv")

# Tính toán số lượng bài báo theo Website source
website_counts = df['Website source'].value_counts()

# Vẽ biểu đồ cho số lượng bài báo theo Website source
plt.figure(figsize=(10, 6))
website_counts.plot(kind='bar')
plt.title('Number of Articles by Website Source')
plt.xlabel('Website Source')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.show()

# Tính toán số lượng bài báo theo Article type
article_type_counts = df['Article type'].value_counts()

# Vẽ biểu đồ cho số lượng bài báo theo Article type
plt.figure(figsize=(8, 5))
article_type_counts.plot(kind='bar')
plt.title('Number of Articles by Article Type')
plt.xlabel('Article Type')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.show()



```

```python
# Chuyển cột Date sang định dạng datetime với đúng định dạng "%d/%m/%Y"
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Tính toán số lượng bài báo theo năm
df['Year'] = df['Date'].dt.year
yearly_counts = df['Year'].value_counts().sort_index()

# Vẽ biểu đồ cho số lượng bài báo theo năm
plt.figure(figsize=(10, 6))
yearly_counts.plot(kind='bar')
plt.title('Number of Articles by Year')
plt.xlabel('Year')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.show()
```

```python
# Tính toán phần trăm số lượng bài báo theo Website source
website_percentage = website_counts / website_counts.sum() * 100

# Vẽ biểu đồ tròn cho phân phối phần trăm số lượng bài báo theo Website source
plt.figure(figsize=(8, 8))
plt.pie(website_percentage, labels=website_percentage.index, autopct='%1.1f%%')
plt.title('Percentage Distribution of Articles by Website Source')
plt.show()
```

```python
# Tính toán phần trăm số lượng bài báo theo Article type
article_type_percentage = article_type_counts / article_type_counts.sum() * 100

# Vẽ biểu đồ tròn cho phân phối phần trăm số lượng bài báo theo Article type
plt.figure(figsize=(8, 8))
plt.pie(article_type_percentage, labels=article_type_percentage.index, autopct='%1.1f%%')
plt.title('Percentage Distribution of Articles by Article Type')
plt.show()
```

```python
# Tính toán số lượng bài báo theo tác giả
author_counts = df['Author\'s'].value_counts()

# Chọn số lượng tác giả hàng đầu (ở đây, chọn 10 tác giả)
top_authors = author_counts.head(30)

# Vẽ biểu đồ tròn cho phân phối số lượng bài báo của các tác giả hàng đầu
plt.figure(figsize=(8, 8))
plt.pie(top_authors, labels=top_authors.index, autopct='%1.1f%%')
plt.title('Percentage Distribution of Articles by Top Authors')
plt.show()
```

```python
# Sắp xếp các tác giả theo số lượng bài báo
sorted_authors = author_counts.sort_values(ascending=False)

# Chọn top 10 tác giả có số bài báo nhiều nhất
top_10_authors = sorted_authors.head(15)

# Vẽ biểu đồ cột cho số lượng bài báo của top 10 tác giả
plt.figure(figsize=(10, 6))
top_10_authors.plot(kind='bar')
plt.title('Number of Articles by Top 10 Authors')
plt.xlabel('Authors')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.show()
```

```python
# Tính toán số lượng bài báo theo Category
category_counts = df['Category'].value_counts()

# Kiểm tra nếu có bài báo nào không có Category thì đưa vào mục "Other"
if 'Other' not in category_counts:
    category_counts['Other'] = len(df) - category_counts.sum()

# Vẽ biểu đồ tròn cho phân phối số lượng bài báo theo Category
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
plt.title('Percentage Distribution of Articles by Category')
plt.show()
```
